```python
from asteroid.models import ConvTasNet
model = ConvTasNet.from_pretrained("mpariente/ConvTasNet_WHAM!_sepclean")
```

```python
import torch
import torchaudio
from IPython.display import Audio

wav, sr = torchaudio.load("/app/src/sample.wav")
wav = torchaudio.transforms.Resample(sr, 8000)(wav)
wav = wav.unsqueeze(0)
print(wav.shape)

separated_sources = model(wav)
print(separated_sources.shape)
```

```python
print('separated')
Audio(separated_sources[0, 0].detach().numpy(), rate = 8000)
```
